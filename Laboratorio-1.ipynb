# Problema de Monty Hall

```python
import numpy as np
```

```python
# N cantidad de casos
N = 1000

def DistribucionPuerta():
    puerta = np.array([0,0,0])
    puerta[np.random.randint(3)] = 1
    return puerta
```

```python
from matplotlib import pyplot as plt
```

```python
ganadas_cambio = 0
cambio_acum = []
ganadas_no_cambio = 0
no_cambio_acum = []
ganadas_cambio_al_azar = 0
cambio_azar_acum = []

for n in range(N):
    puertas = DistribucionPuerta()
    puerta_elegida = np.random.randint(3)
    
    #Supongo que cambio siempre
    if(puertas[puerta_elegida]==0): #Cambió y la puerta que habia elegido era 0, entonces GANÓ!
        ganadas_cambio += 1
    cambio_acum.append(ganadas_cambio)
    
    #Supongo que no cambio nunca
    if(puertas[puerta_elegida]==1): #No cambió de puerta, y la que había elegido es 1, entonces GANÓ! 
        ganadas_no_cambio += 1
    no_cambio_acum.append(ganadas_no_cambio)
    
    #Supongo que cambio de manera aleatoria
    cambio = np.random.randint(2)
    if(cambio == 1):
        if(puertas[puerta_elegida]==0): #Cambió y GANÓ!
            ganadas_cambio_al_azar += 1
    elif(puertas[puerta_elegida]==1): #No cambió y GANÓ!
            ganadas_cambio_al_azar += 1
    cambio_azar_acum.append(ganadas_cambio_al_azar)
    
print('Aciertos con cambio: ', ganadas_cambio)
print('Aciertos sin cambio: ', ganadas_no_cambio)
print('Aciertos con cambio al azar: ', ganadas_cambio_al_azar)

plt.figure(figsize=(16,8))
plt.xlabel('Casos')
plt.ylabel('Victorias')
plt.title('Monty Hall')
plt.plot(no_cambio_acum,label='No Cambio')
plt.plot(cambio_acum,label='Cambio')
plt.plot(cambio_azar_acum,label='Al Azar')
plt.legend()
plt.show()
```
