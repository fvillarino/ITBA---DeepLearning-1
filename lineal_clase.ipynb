```python
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
x = np.array([0,20,50])
y = np.array([-3,37,97])
plt.scatter(x,y)
```

```python
lr = LinearRegression()
lr.fit(x.reshape(-1,1),y.reshape(-1,1))
lr.coef_, lr.intercept_
```
