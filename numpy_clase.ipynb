# Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
x = [1,4,6]
y = [2,5,7]
x2 = [1,3,5,1,3]
y2 = [1,2,3,5,2]
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0, 7, 0, 7])
plt.title('Gr√°fico')
plt.plot(x,y, label='Blue line')
plt.plot(x2,y2,'ro', label='Other line')
plt.legend()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
```

```python
x=np.arange(0,5,1)
y=x**2
y1=x**3
y2=1/x
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y,x,y1,x,y2)
```
